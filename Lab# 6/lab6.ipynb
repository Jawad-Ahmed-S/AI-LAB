**QUESTION # 1**

```python
import random

def f(x):
  return -x**2 + 6*x

def hillClimbing(start):
  currX = start
  print("Initial: ",currX)

  while True:
    currF = f(currX)
    print(f"Current x = {currX}, f(x) = {currF}")

    nextX = currX+1
    prevX = currX-1


    if nextX <= 6:
      nextF = f(nextX)
    else:
      nextF = -float('inf')

    if nextX >= 0:
      prevF = f(prevX)
    else:
      prevF = -float('inf')

    if nextF> currF and nextF >= prevF:
      currX = nextX
    elif prevF> currF and prevF >= prevF:
      currX = prevX
    else:
      return currX,currF

start = random.randint(0,6)
finalX,finalF = hillClimbing(start)

print("\n\nFinal x: ",finalX,"\nFinal F(x): ",finalF)
```

**QUESTION # 2**

```python
def h(n):
  return abs(20-n)
beam = [1]

while 20 not in beam:
  neighbours = []

  for x in beam:
    neighbours.append(x+2)
    neighbours.append(x+3)
    neighbours.append(x*2)

  neighbours.sort(key=h)
  selectedNeighbours = []
  for n in neighbours:
    if n not in selectedNeighbours:
      selectedNeighbours.append(n)
    if len(selectedNeighbours) ==2 :break

  beam = selectedNeighbours
  print(f"Current best states: {beam}")


print("Goal 20 reached!")
```
