```python
class Vehicle:
  def __init__(self,id,brand,rent_per_day):
    self.__id =id
    self.__brand = brand
    self.__rent_per_day = rent_per_day
  def display_details(self):
      print("ID: ",self.__id)
      print("Brand: ",self.__brand)
      print("Rent: ",self.__rent_per_day)
      print("\n")
  def calculateRent(self,days):
      print("Rent for ",days," days is :",self.__rent_per_day*days)



v1 = Vehicle(1,"Toyota",20000)
v2 = Vehicle(3,"Kia",25000)

v1.display_details()
v2.display_details()

v1.calculateRent(5)
v2.calculateRent(6)
```

```python
class Employee:
  def __init__(self,id,name):
    self._id = id
    self._name = name


class fullTimeEmployee(Employee):
  def __init__(self,id,name,monthlySalary):
    self.__monthlySalary = monthlySalary
    super().__init__(id,name)
  def calculateSalary(self):
    print("Name :",self._name)
    print("Salary :",self.__monthlySalary)
    print("\n")

class partTimeEmployee(Employee):
  def __init__(self,id,name,hoursWorked,hourlyRate):
    self.__hoursWorked = hoursWorked
    self.__hourlyRate = hourlyRate
    super().__init__(id,name)
  def calculateSalary(self):
    print("Name :",self._name)
    print("Salary :",self.__hoursWorked * self.__hourlyRate)
    print("\n")


f1 = fullTimeEmployee(1,"Mr. A",10000)
p2 = partTimeEmployee(2,"Mr. B",40,100)

f1.calculateSalary()
p2.calculateSalary()

```

```python
class Student:
  def __init__(self,id,name,marks):
    self.__id = id
    self.__name = name
    self.__marks = marks
  def setMarks(self,marks):
    self.__marks =marks
  def getMarks(self):
    return self.__marks
  def calculateGrade(self):
    if self.__marks >= 80:
      return "A"
    elif self.__marks >= 70:
      return "B"
    elif self.__marks >= 60:
      return "C"
    elif self.__marks >= 50:
      return "D"
    else:
      return "Fail!"

st1 = Student(1,"Mr. A",50)
st2 = Student(2,"Mr. B",75)

st1.setMarks(90)
st1Grade =st1.calculateGrade()
st2Grade =st2.calculateGrade()

print("Student 1 Grade: ",st1Grade)
print("Student 2 Grade: ",st2Grade)
```
