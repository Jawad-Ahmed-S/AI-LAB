```python
class enviornment:
  def __init__(self,state):
    self.state=state
  def getState(self):
    return self.state
  def updateEnviornment(self,action):
    if action =='Heavy Traffic':
      self.state = 'Heavy Traffic'
    else :
      self.state = 'Light Traffic'

class SimpleReflexAgent:
  def ChooseAction(self,percept):
    if percept == 'Heavy Traffic':
      return 'Extend Green Light'
    else:
      return 'Normal Green Light'
def runAgent(percept):
  env = enviornment(percept)
  agent = SimpleReflexAgent()
  action = agent.ChooseAction(percept)
  env.updateEnviornment(action)
  print("Percept: ",percept,"-> Action: ",action)
  return action,env.getState()


runAgent("Heavy Traffic");
runAgent("Light Traffic");
```

```python
class enviornment:
  def __init__(self,studentsPresent,LigthStatus):
    self.studentsPresent =studentsPresent
    self.LigthStatus =LigthStatus

  def getState(self):
    return self.studentsPresent,self.LigthStatus

  def updateEnviornment(self,action):
    if action =='On':
      self.LigthStatus = 'On'
    else :
      self.LigthStatus = 'Off'

class ModelReflexAgent:
  def __init__(self):
    self.studentsPresent =None
    self.LigthStatus =None

  def updateModel(self,studentsPresent,LigthStatus):
    self.studentsPresent = studentsPresent
    self.LigthStatus = LigthStatus

  def ChooseAction(self):
    if self.studentsPresent== 'Yes' and self.LigthStatus == 'Off':
      return 'On'
    elif self.studentsPresent== 'No' and self.LigthStatus == 'On':
      return 'Off'
    else:
      return 'No Action'



env = enviornment("Yes","Off")
agent = ModelReflexAgent()

for i in range(8):
  percept = env.getState()
  agent.updateModel(percept[0],percept[1])
  action = agent.ChooseAction()
  env.updateEnviornment(action)
  print("Percept: ",percept[0]," AND ",percept[1],"-> Action: ",action)


```

```python
class enviornment:
  def __init__(self):
    self.subjects = {
        "AI":False,
        "Math":False,
        "Physics":False
    }

  def getState(self):
    return self.subjects.copy()

  def updateEnviornment(self,subject):
    if subject in self.subjects:
      self.subjects[subject]= True

class GoalBasedAgent:
  def __init__(self):
    self.internalModel ={}

  def updateModel(self,percept):
    self.internalModel = percept

  def isGoalAcheived(self):
    for status in self.internalModel.values():
      if not status:
        return False
    return True


  def ChooseAction(self):
    for subject in self.internalModel:
      if self.internalModel[subject] == False:
        return subject
    return 'No Action'


env = enviornment()
agent = GoalBasedAgent()

while True:
  percept = env.getState()
  agent.updateModel(percept)

  if agent.isGoalAcheived():
        print("Goal Achieved")
        break

  action = agent.ChooseAction()
  print("Studying: ",action)
  env.updateEnviornment(action)


```
